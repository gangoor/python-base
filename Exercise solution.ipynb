```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Telecom_data.csv")
```

```python
print(df.columns)
```

```python
# 1. Married vs unmarried people, what is the average of income?

df[["marital","income"]].groupby("marital").mean()
```

```python
# 2. What is the average years of employment?

df.employ.mean()
```

```python
# 3. Is there a difference in number of retired people as a 
# percentage across different regions?
df[["region","retire"]].groupby("region").sum().apply(lambda x:x/x.sum())
```

```python
# 4. Does "ed" education, show any relationship to income?
df[["ed","income"]].groupby("ed").mean()
# increasing income with increasing ed observed
```

```python
# 5. Does gender have any relationship with the usage of tollfree service?
df[["gender","tollfree"]].groupby("gender").agg(["sum","count"])
# doesnt appear to be a big difference
```

```python
# 6. List all customers who have a tenure of more than 15
df.loc[df.tenure > 15].head()
```

```python
# 7
df[["ed","gender"]].sort_values(by=["ed","gender"],ascending=[False,False])
```

```python
# 8. Which customer category (custcat) has the most number of churn? 
# (churn = people who switch telecom providers.)

df[["custcat","churn"]].groupby("custcat").agg(['sum', 'count'])
```

```python
# 9. Do customers who opt for ebill have lesser propensity to churn?
df[["ebill","churn"]].groupby("ebill").agg(['sum', 'count'])
```
