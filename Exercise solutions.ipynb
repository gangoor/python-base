## 6. Write a function that calculates "dob" given age and a  random number between 1 and 365.
 - If age is 20, and a random number if 180, then dob = "7-10-2020" - (20 years + 180 days)
 - (This will require you to understand how to handle dates, research for an answer independently)

```python
from datetime import datetime, timedelta


def dob(x):
    days_to_subtract = x*365 + random.randint(0,364)
    d = datetime.today() - timedelta(days=days_to_subtract)
    return d

DF4["dob"] = DF4["age"].apply(dob)
```
